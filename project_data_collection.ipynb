Libs for collecting

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd 
```

Collecting the Urls

```python
base_url = 'https://mushokutensei.fandom.com/wiki/Category:Characters?from='

all_urls = []

for i in range(0,26):
    letter=chr(ord('A')+i)
    print(letter)
    driver.get(base_url+letter)
    elements = driver.find_elements(by=By.CLASS_NAME,value='category-page__member-link')
    urls=[]
    for element in elements:
        url=element.get_attribute('href')
        print(url)
        urls.append(url)
        print(url)
    all_urls.append(urls)
```

Collecting the data from the urls

```python
chrome_options = Options()
chrome_options.add_argument("--headless")  # Activate headless mode
chrome_options.add_argument("--disable-gpu")  # Disable GPU acceleration (often needed in headless mode)

driver = webdriver.Chrome(options=chrome_options)

citizens=[]
for char_page_url in all_urls:
    print(char_page_url)
    try:
        driver.get(char_page_url)
        name_of_a_charachter = driver.find_element(by=By.CLASS_NAME,value='mw-page-title-main').text
        print(name_of_a_charachter)
        names_of_the_fields=driver.find_elements(by=By.CLASS_NAME,value='pi-data-label')
        field_values=driver.find_elements(by=By.CLASS_NAME,value='pi-data-value')
        citizen = {}
        citizen['Name']=name_of_a_charachter
        for i in range(len(names_of_the_fields)):
            print(names_of_the_fields[i].text,field_values[i].text)
            citizen[names_of_the_fields[i].text]=field_values[i].text
        citizens.append(citizen)
    except Exception as e:
        print("yo theres been an error")
```

Creating a dataframe

```python
df = pd.DataFrame(citizens)
print(df.shape)
df
```

Saving

```python
df.to_csv('all_characters_in_mushoku_tensei.csv', index=False)
```
